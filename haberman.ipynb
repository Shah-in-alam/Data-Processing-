```python
median =np.median(haberman["1"])
print(median)
```

```python
plt.hist(haberman["30"], bins = 50, alpha = 0.8)
plt.axvline(haberman["30"].mean(), 0, 150, label = "mean", color = "red")
plt.axvline(haberman["30"].median(), 0, 150, label = "median", color = "orange")
plt.legend()
plt.show()
```

```python
variance =np.var(haberman["30"])
stdev= np.sqrt(variance)
print(stdev)
```
